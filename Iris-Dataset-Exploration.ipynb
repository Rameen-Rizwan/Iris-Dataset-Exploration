### Objective 3: Data Visualization

```python
#Libraries
import seaborn as sns
import matplotlib.pyplot as plt
```

#### Scatter PLot

```python
#To analyze relationships between variables
sns.scatterplot(data=df, x='sepal_length', y='petal_length', hue='species')
plt.title("Sepal Length vs Petal Length")
plt.show()
```

#### Histogram

```python
#To examine data distribution
sns.histplot(df['sepal_length'], kde=True)
plt.title("Distribution of Sepal Length")
plt.show()
```

#### Box Plot

```python
#To detect outliers and spread of values
sns.boxplot(data=df, x='species', y='petal_length')
plt.title("Box Plot of Petal Length by Species")
plt.show()
```

## Conclusion:

- The Iris dataset includes 150 records with 5 main attributes related to sepal and petal measurements and species classification.
- Each Iris species shows distinct patterns in petal and sepal dimensions, which can aid in classification.
- Scatter plots highlighted clear relationships between features such as sepal length and petal length.
- Histograms provided insights into the distribution and frequency of values across features.
- Box plots helped identify outliers and understand the spread and central tendency of the data.
- The exploratory analysis offered a strong understanding of the dataset's structure and key feature behavior.
